```python
import numpy as np
import pandas as pd

df = pd.read_csv("manual_annotation_task_ML4H 2.csv")
df.head()
```

```python
# Lets get the Cohen's Kappa Score for the dataset
from sklearn.metrics import cohen_kappa_score

first_annotator = df['annotator1']
second_annotator = df['annotator2']

# Remove rows with missing values
valid_mask = first_annotator.notna() & second_annotator.notna()
kappa_score = cohen_kappa_score(first_annotator[valid_mask], second_annotator[valid_mask])

print(f"Cohen's Kappa Score - Annotator 1 vs Annotator 2: {round(kappa_score, 2)}")

# print(len(first_annotator.notna()))
# print(len(second_annotator.notna()))
```

```python
# Compare each annotator's labels vs default labels

default_status = df['status']

kappa_annotator1 = cohen_kappa_score(first_annotator[valid_mask], default_status[valid_mask])
kappa_annotator2 = cohen_kappa_score(second_annotator[valid_mask], default_status[valid_mask])

print(f"Cohen's Kappa Score for Annotator 1 vs Default: {round(kappa_annotator1, 2)}")
print(f"Cohen's Kappa Score for Annotator 2 vs Default: {round(kappa_annotator2, 2)}")
```
